${\bf Problema}$ ${\bf 1:}$ Use o Python para plotar os campos de direção para a família de souluções de cada PVI do problema 1 da Aula 1.



```python
#----------------- Exercicio 1 -----------------
#------------ EDO: y'=-[y(t)]^2+y(t), y(0)=1 -------------

import numpy as np
import matplotlib.pyplot as plt

# parametros
t0 , tb = 0, 2
y0 , yb = 1, 2
nt , ny = 12, 18

# dominio (t,y)
t = np. linspace (t0 ,tb ,nt)
y = np. linspace (y0 ,yb ,ny)
[T,Y] = np.meshgrid (t,y)

# EDO
dt = np.ones (T.shape)
dy = -(Y)**2 + Y

# campo
f = plt.figure (figsize = (6,5))
plt.quiver (T, Y, dt, dy, color ='g')
```

${\bf Problema}$ ${\bf 2:}$ Plotar as soluções particulares substituindo ${c}$

```python
#------------- Problema 1 ---------------
#----------- EDO: y'=-y+sin(t)+cos(t), y(0)=1 -----------

import numpy as np
import matplotlib.pyplot as plt

# parametros
t0 , tb = 0, 2
y0 , yb = -2, 3
nt , ny = 12, 18

# dominio (t,y)
t = np.linspace (t0 ,tb ,nt)
y = np.linspace (y0 ,yb ,ny)
[T,Y] = np.meshgrid (t,y)

# EDO
dt = np.ones (T.shape)
dy = -Y + np.sin(T) + np.cos(T)

# campo
f = plt.figure ( figsize =(6,5) )
plt.quiver (T, Y, dt, dy, color = 'c')

# solucoes particulares
y_part = lambda c: np.sin(t) + c*np.exp(-t)

for c in [-3., -2., 2., 3.]:
    aux = y_part(c)
    plt.plot (t, y_part (aux))
```

```python
#------------- Problema 2 ---------------
# ----------- EDO: y'=(y-y^2)/t, y(1)=2 ---------------------

import numpy as np
import matplotlib.pyplot as plt

# parametros
t0 , tb = 1, 3
y0 , yb = -2, 2
nt , ny = 12, 18

# dominio (t,y)
t = np.linspace (t0 ,tb ,nt)
y = np.linspace (y0 ,yb ,ny)
[T,Y] = np.meshgrid (t,y)

# EDO
dt = np.ones (T.shape)
dy = (Y-Y**2)/T

# campo
f = plt.figure ( figsize =(6,5) )
plt.quiver (T, Y, dt, dy, color = 'c')

# solucoes particulares
y_part = lambda c: t/(t+c)

for c in [-1, -2, 2, 1.]:
    aux = y_part(c)
    plt.plot (t, y_part (aux))
```
